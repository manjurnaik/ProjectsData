```python
sn.barplot(x='Gender',y='Amount',data=df_gp)
```

```python
bx=sn.countplot(data=df,x='Age Group',hue='Gender')
for bars in bx.containers:
    bx.bar_label(bars)
```

```python
sales_age_df=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sn.barplot(x='Age Group',y='Amount',data=sales_age_df)
```

```python
sales_state_df=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sn.set(rc={'figure.figsize':(25,5)})
sn.barplot(x='State',y='Orders',data=sales_state_df)
```
